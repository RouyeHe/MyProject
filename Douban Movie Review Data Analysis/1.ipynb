```python
import pandas as pd
df = pd.read_csv('DMSC.csv')
```

```python
print(df.columns)
```

```python
# 筛选出特定的列
df = df[['Movie_Name_CN', 'Number','Username', 'Date','Star', 'Comment', 'Like']]
```

```python
df1 = df[df['Movie_Name_CN'] =="复仇者联盟"]
```

```python
df1.head()
```

```python
from pyecharts.charts import Bar
from pyecharts import options as opts

# 对每个评分等级的评论计算平均点赞数和评论数
grouped_data = df1.groupby('Star').agg({'Like': 'mean', 'Comment': 'size'}).reset_index()

# 创建柱状图
bar = (
    Bar()
    .add_xaxis(xaxis_data=list(grouped_data['Star'].astype(str)))
    .add_yaxis(
        series_name="平均点赞数",
        y_axis=list(grouped_data['Like']),
        yaxis_index=0,
        label_opts=opts.LabelOpts(is_show=False),
    )
    .add_yaxis(
        series_name="评论数",
        y_axis=list(grouped_data['Comment']),
        yaxis_index=1,
        label_opts=opts.LabelOpts(is_show=False),
    )
    .extend_axis(
        yaxis=opts.AxisOpts(
            name="评论数",
            type_="value",
            position="right",
            axislabel_opts=opts.LabelOpts(formatter="{value}"),
        )
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="各评分的评论的平均点赞数与评论数"),
        yaxis_opts=opts.AxisOpts(
            name="平均点赞数",
            type_="value",
            axislabel_opts=opts.LabelOpts(formatter="{value}"),
        ),
        xaxis_opts=opts.AxisOpts(name="评分"),
    )
)

# 渲染图表到文件
bar.render("各评分的评论数据分布.html")
```

```python
from pyecharts.charts import Line
from pyecharts import options as opts
# 确保Date列是datetime类型
df1['Date'] = pd.to_datetime(df1['Date'])

# 按日期分组并计算平均点赞数
grouped_data = df1.groupby('Date').agg({'Star': 'mean'}).reset_index()

# 转换日期格式为字符串，以便在图表中显示
grouped_data['Date'] = grouped_data['Date'].dt.strftime('%Y-%m-%d')

# 创建线图
line_chart = (
    Line()
    .add_xaxis(xaxis_data=list(grouped_data['Date']))
    .add_yaxis(
        series_name="平均点赞数",
        y_axis=list(grouped_data['Star']),
        label_opts=opts.LabelOpts(is_show=False),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="每个日期的平均点赞数变化趋势"),
        xaxis_opts=opts.AxisOpts(type_="category", name="日期"),
        yaxis_opts=opts.AxisOpts(name="平均评分"),
        tooltip_opts=opts.TooltipOpts(trigger="axis"),
    )
)

# 渲染图表到HTML文件中
line_chart.render("每个日期评论的平均评分变化趋势.html")
```

```python
from pyecharts.charts import Line
from pyecharts import options as opts

# 确保Date列是datetime类型
df1['Date'] = pd.to_datetime(df1['Date'])

# 按月份分组并计算平均评分
grouped_data = df1.groupby(df1['Date'].dt.to_period('M')).agg({'Star': 'mean'}).reset_index()

# 将日期从Period转换回字符串格式，以便在图表中显示
grouped_data['Date'] = grouped_data['Date'].dt.strftime('%Y-%m')

# 创建线图
line_chart = (
    Line()
    .add_xaxis(xaxis_data=list(grouped_data['Date']))
    .add_yaxis(
        series_name="平均评分",
        y_axis=list(grouped_data['Star']),
        label_opts=opts.LabelOpts(is_show=False),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="每个月评论的平均评分变化趋势"),
        xaxis_opts=opts.AxisOpts(type_="category", name="月份"),
        yaxis_opts=opts.AxisOpts(name="平均评分"),
        tooltip_opts=opts.TooltipOpts(trigger="axis"),
    )
)

# 渲染图表到HTML文件中
line_chart.render("每个月评论的平均评分变化趋势.html")
```

```python
# 确保数据按日期排序
df1.sort_values(by='Date', inplace=True)

# 初始化一个空的DataFrame来存储结果
influence_df = pd.DataFrame(columns=['Top_Comment_Index', 'Average_Star_After'])

# 遍历点赞数前五的评论
for index, top_comment in top_likes.iterrows():
    # 找出当前评论之后的所有评论
    subsequent_reviews = df1[df1['Date'] > top_comment['Date']]
    
    # 计算这些评论的平均评分
    average_star_after = subsequent_reviews['Star'].mean() if not subsequent_reviews.empty else None
    
    # 将结果存储在DataFrame中
    influence_df = influence_df.append({
        'Top_Comment_Index': index,
        'Average_Star_After': average_star_after
    }, ignore_index=True)
```

```python
import pandas as pd
from pyecharts.charts import Bar3D
from pyecharts import options as opts

# 计算每部电影的平均点赞数和平均评分
movie_grouped = df.groupby('Movie_Name_CN').agg({'Like': 'mean', 'Star': 'mean'}).reset_index()

# 构造三维数据
data = [
    [i, movie_grouped.loc[i, 'Star'], movie_grouped.loc[i, 'Like']]
    for i in range(movie_grouped.shape[0])
]

# 创建3D柱状图
bar3d = (
    Bar3D()
    .add(
        series_name="",
        data=data,
        xaxis3d_opts=opts.Axis3DOpts(type_="category", data=movie_grouped['Movie_Name_CN'].tolist(), name='电影'),
        yaxis3d_opts=opts.Axis3DOpts(type_="value", name='平均评分'),
        zaxis3d_opts=opts.Axis3DOpts(type_="value", name='平均点赞数'),
    )
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(max_=movie_grouped['Like'].max()),
        title_opts=opts.TitleOpts(title="电影评论的平均点赞数与平均评分")
    )
)

# 渲染图表到HTML文件中
bar3d.render("平均点赞数,平均评分,电影三维.html")
```

```python
from pyecharts.charts import Bar3D
from pyecharts import options as opts

# 计算每部电影的总评论数和平均评分
movie_grouped = df.groupby('Movie_Name_CN').agg({'Like': 'sum', 'Star': 'mean', 'Comment': 'count'}).reset_index()

# 构造三维数据，其中x轴是电影索引，y轴是平均评分，z轴是总评论数

data = [
    [i, movie_grouped.loc[i, 'Star'], movie_grouped.loc[i, 'Comment'] / 10000]  # 这里将总评论数除以10000
    for i in range(movie_grouped.shape[0])
]
max_comment_scaled = movie_grouped['Comment'].max() / 10000

# 创建3D柱状图
bar3d = (
    Bar3D()
    .add(
        series_name="",
        data=data,
        xaxis3d_opts=opts.Axis3DOpts(type_="category", data=movie_grouped['Movie_Name_CN'].tolist(), name='电影'),
        yaxis3d_opts=opts.Axis3DOpts(type_="value", name='平均评分'),
        zaxis3d_opts=opts.Axis3DOpts(type_="value", name='总评论数/万条')
    )
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(max_=max_comment_scaled),
        title_opts=opts.TitleOpts(title="电影评论数量、平均评分与电影的三维关系图")
    )
)

# 渲染图表到HTML文件中
bar3d.render("总评论数_平均评分_电影三维关系图.html")
```
